```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
file_df = pd.read_csv(file, encoding="ISO-8859-1")
file_df.head(5)
```

```python
# Inspect all columns
file_df.count()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
coders_df = file_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
coders_df
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
coders_df = coders_df.replace({0:"No"})
coders_df = coders_df.replace({1:"Yes"})
coders_df
```

```python
#Calculate total number of respondents in survey
index = coders_df.index
respondents = len(index)
respondents 
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_df = coders_df.loc[coders_df["AttendedBootcamp"] == "Yes"]
attended_df
```

```python
coders_df['Gender'].value_counts()
```

```python
# Calculate average age of attendees
avgage = attended_df["Age"].mean()
print("Average Age of Attendees", round(avgage, 2))
# Calculate how many people attended a bootcamp
attended_index = attended_df.index
attended = len(attended_index)
print("BootCamp Attendees:", attended)
# Calculate how many attendees hold degrees
holds_degree = attended_df["SchoolDegree"].loc[
    (attended_df["SchoolDegree"] != "some college credit, no degree") & 
    (attended_df["SchoolDegree"] != "high school diploma or equivalent (GED)") &
    (attended_df["SchoolDegree"] != "trade, technical, or vocational training") &
    (attended_df["SchoolDegree"] != "some high school") &
    (attended_df["SchoolDegree"] != "no high school (secondary school)")
     ].count()
print('Attendees that hold degrees:', holds_degree)
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
gender = attended_df['Gender'].value_counts()
print(gender)
# Calculate percentage of respondents who attended a bootcamp
percent = (attended/respondents) * 100
print('Percentage of respondents who attended a bootcamp:',round(percent,3), "%")
# Calculate percentage of respondents belonging to each gender
male = coders_df['Gender'].loc[(coders_df['Gender'] == 'male')].count()
per_male = (male / respondents) * 100
female = coders_df['Gender'].loc[(coders_df['Gender'] == 'female')].count()
per_female = (female / respondents) * 100
gq = coders_df['Gender'].loc[(coders_df['Gender'] == 'genderqueer')].count()
per_gq =(gq / respondents) * 100
ag = coders_df['Gender'].loc[(coders_df['Gender'] == 'agender')].count()
per_ag = (ag / respondents) * 100
trans = coders_df['Gender'].loc[(coders_df['Gender'] == 'trans')].count()
per_trans = (trans / respondents) * 100
print('Percent of male respondents =', round(per_male, 3), '%')
print('Percent of female respondents =', round(per_female,3), '%')
print('Percent of genderqueer respondents =', round(per_gq,3), '%')
print('Percent of agender respondents =', round(per_ag,3), '%')
print('Percent of trans respondents =', round(per_trans,3), '%')
per_nonsp = per_ag + per_gq + per_trans
# Calculate percentage of attendees with a school degree
per_degree = (holds_degree/attended) * 100 
print('Percentage of attendees with degrees =', round(per_degree, 2), '%')
# Calculate average post-bootcamp salary
av_salary = attended_df["BootcampPostSalary"].mean()
print('Average Post-BootCamp Salary: $', round(av_salary, 3))
```

```python
# Create a new table consolodating above calculations
bootcamp_breakdown = pd.DataFrame({
    "Total Surveyed": [respondents], 
    "Total Bootcamp Attendees" : [attended], 
    "% Attended Bootcamp" : [percent], 
    "Avg. Age" : [avgage], 
    "% Male" : [per_male], 
    "% Female": [per_female], 
    "% Non Gender Specific": [per_nonsp], 
    "Has a Degree": [per_degree],
    "Average Post Bootcamp Salary": [av_salary]})
bootcamp_breakdown = bootcamp_breakdown[["Total Surveyed", 
                                         "Total Bootcamp Attendees", 
                                         "% Attended Bootcamp", 
                                         "Avg. Age", 
                                         "Has a Degree", 
                                         "% Male", 
                                         "% Female", 
                                         "% Non Gender Specific", 
                                         "Average Post Bootcamp Salary"]]
bootcamp_breakdown
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_breakdown["% Attended Bootcamp"] = bootcamp_breakdown["% Attended Bootcamp"].map("{0:,.2f}%".format)
bootcamp_breakdown["Avg. Age"] = bootcamp_breakdown["Avg. Age"].map("{0:,.0f}".format)
bootcamp_breakdown["Has a Degree"] = bootcamp_breakdown["Has a Degree"].map("{0:,.2f}%".format)
bootcamp_breakdown["% Male"] = bootcamp_breakdown["% Male"].map("{0:,.2f}%".format)
bootcamp_breakdown["% Female"] = bootcamp_breakdown["% Female"].map("{0:,.2f}%".format)
bootcamp_breakdown["% Non Gender Specific"] = bootcamp_breakdown["% Non Gender Specific"].map("{0:,.2f}%".format)
bootcamp_breakdown["Average Post Bootcamp Salary"] = bootcamp_breakdown["Average Post Bootcamp Salary"].map("${0:,.0f}".format)
bootcamp_breakdown
```

```python
# Export to Excel
bootcamp_breakdown.to_excel("output/BootcampOutputPart1.xlsx", index=False)
```
