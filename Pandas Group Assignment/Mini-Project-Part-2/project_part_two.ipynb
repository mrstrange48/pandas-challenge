```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
file_df = pd.read_csv(file, encoding="ISO-8859-1")
file_df.head(5)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
coders_df = file_df.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
coders_df
```

```python
# Change reading 0 and 1 to No and Yes, respectively
coders_df = coders_df.replace({0:"No"})
coders_df = coders_df.replace({1:"Yes"})
coders_df
```

```python
# Extract rows for only those who attended a bootcamp
attended_df = coders_df.loc[coders_df["AttendedBootcamp"] == "Yes"]
attended_df
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
attended_df['BootcampName'].value_counts()
#defining bootcamps_count as equal to the value counts of the 'BootcampName' column
bootcamps_count = attended_df['BootcampName'].value_counts(dropna=True, sort=True)
#setting the value count as the data frame 
bootcampcount_df = pd.DataFrame(bootcamps_count)
bootcamps_df = bootcampcount_df.reset_index()
#change name of the columns
bootcamps_df.columns = ['Bootcamp', 'Attendees']
bootcamps_df
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend_df = attended_df.loc[attended_df['BootcampRecommend'] == 'Yes']
rec_bootcamp_count = recommend_df['BootcampName'].value_counts()
rec_count_df = pd.DataFrame(rec_bootcamp_count)
rec_df = rec_count_df.reset_index()
rec_df.columns = ['Bootcamp', 'Recommend']
rec_df
```

```python
# Merge the two created data frames on the name of the bootcamp
merged_df = pd.merge(bootcamps_df, rec_df)
merged_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders

merged_df['% Recommended']= round((merged_df['Recommend'] / merged_df['Attendees']) * 100, 2)
merged_df
```

```python
# Export to excel and remove index
merged_df.to_excel("output/BootcampOutputPart2.xlsx", index=False)
```
